# restnet

```python
from tensorflow.keras.applications.resnet50 import preprocess_input as RN50_preprocess_input
from tensorflow.keras.applications.resnet50 import ResNet50
```

```python
train_datagen = ImageDataGenerator(
            rotation_range = 20,  # randomly rotate images in the range (degrees, 0 to 180)
            width_shift_range = 0.2,  # randomly shift images horizontally (fraction of total width)
            height_shift_range = 0.2,  # randomly shift images vertically (fraction of total height)
            shear_range = 0.2,
            zoom_range = 0.2, # Randomly zoom image 
            horizontal_flip = True,  # randomly flip images
            fill_mode = 'nearest',
            preprocessing_function = RN50_preprocess_input)
test_datagen = ImageDataGenerator(preprocessing_function = RN50_preprocess_input)
```

```python
train_generator = train_datagen.flow(x=x_train, y=y_train)
test_generator = test_datagen.flow(x=x_test, y=y_test)
```

# transfer learning

```python
pre_trained_RN50 = ResNet50(include_top=False, weights='imagenet', input_shape=(224,224,3))
pre_trained_RN50.trainable = False
pre_trained_RN50.summary()
```

```python
tf.keras.backend.clear_session() 

RN50 = models.Sequential([pre_trained_RN50,
                        layers.Flatten(),
                        layers.Dropout(0.3),
                        layers.Dense(4, activation = "softmax")])
```

```python
# Compile the model
RN50.compile(optimizer=optimizers.Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
RN50.summary()
```

```python
# Train the model
history=RN50.fit(train_generator, epochs=10, validation_data=test_generator, batch_size=32)
```

```python
RN50.evaluate(test_generator)
```

```python
# Save the model to a file
RN50.save('RN50_model.h5')
```

```python
# Plot of model comparison

barWidth=0.2

r1 = np.arange(len(precision))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

fig, ax = plt.subplots(figsize=(15, 10))

ax.bar(r1, precision, width=barWidth, label='Precision')
ax.bar(r2, recall, width=barWidth, label='Recall')
ax.bar(r3, f1_score, width=barWidth, label='F1-score')
ax.bar(r4, accuracy, width=barWidth, label='Accuracy')

ax.set_xticks([r + barWidth for r in range(len(precision))])
ax.set_xticklabels(models, fontsize=15)

ax.set_title('Performance of Different Models')
ax.set_xlabel('Models')
ax.set_ylabel('Metrics')

# Add a legend and show the plot
ax.legend(fontsize=12)
plt.show()
```
